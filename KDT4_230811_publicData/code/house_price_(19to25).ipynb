```python
import pandas as pd
import numpy as np
```

```python
house_size_DF=pd.read_excel('../data/지역별_규모별_㎡당_평균_분양가격_20230810225034.xlsx',header=[0,1])
house_size_DF.head()
```

```python
house_size_DF_copy=house_size_DF.copy()
```

```python
house_size_DF_copy1 = house_size_DF_copy.drop(columns=('지역별(1)','지역별(1)')).set_index(('지역별(2)','지역별(2)')).loc[['서울','경기','대구']].T.reset_index().set_index('level_1').loc['전용면적 60㎡초과 85㎡이하']
house_size_DF_copy1
```

```python
house_size_DF_copy1['연도']=house_size_DF_copy1['level_0'].str[0:4]
house_size_DF_copy1
```

```python
house_size_DF_copy1.set_index('연도',drop=True,inplace=True)
```

```python
house_size_DF_copy1.drop(columns='level_0',axis=1,inplace=True)
```

```python
# 천 단위 데이터 만단위로 수정
house_size_DF_copy2=house_size_DF_copy1.groupby('연도').mean()/10
house_size_DF_copy2
```

```python
# 증가율 계산
variation=pd.DataFrame(columns=['서울특별시','경기도','대구광역시'])
for i in range(2015,2024):
    if i < 2023:
        variation.loc[f'{i}~{i+1}']=((house_size_DF_copy2.loc[f'{i+1}']-house_size_DF_copy2.loc[f'{i}'])/house_size_DF_copy2.loc[f'{i}']).to_list()

variation
```

```python
ratio=variation.mean()+1
ratio
```

```python
# 60초과 85이하(18.4525평~ 25.7125평) => 19~25평
# 1m^2 당 금액
# 19평으로 데이터 출력 확인
size=19
finaldata=house_size_DF_copy1.groupby('연도').mean()/100000*size*3.3
```

```python
finaldata_copy=finaldata.copy()
```

```python
finaldata_copy
```

```python
for i in range(2023,2054):
    if i < 2054:
        finaldata_copy.loc[f'{i+1}']=(finaldata_copy.loc[f'{i}']*ratio.to_list())
finaldata_copy.loc['2040':'2054']
```

```python
# 컬럼명 통일(서울->서울특별시, 경기->경기도, 대구->대구광역시)
finaldata_copy.columns=['서울특별시','경기도','대구광역시']
```

```python
finaldata_copy.reset_index(inplace=True)
finaldata_copy.head()
```

```python
finaldata_copy['연도']=list(map(int,finaldata_copy.reset_index()['연도']))
```

```python
finaldata_copy.set_index('연도',inplace=True)
```

```python
finaldata_copy.head()
```

```python
# 60초과 85이하(18.4525평~ 25.7125평) => 19~25평
# 1m^2 당 금액
# csv 파일로 생성
for size in range(19,26):
    finaldata=house_size_DF_copy1.groupby('연도').mean()/100000*size*3.3
    finaldata_copy=finaldata.copy()
    finaldata_copy.reset_index(inplace=True)
    # finaldata_copy.drop(columns='index',inplace=True)
    finaldata_copy['연도']=list(map(int,finaldata_copy.reset_index()['연도']))
    finaldata_copy.set_index('연도',inplace=True)
    finaldata_copy.index
    for i in range(2023,2054):
        if i < 2054:
            finaldata_copy.loc[i+1]=(finaldata_copy.loc[i]*ratio.to_list())
    finaldata_copy.columns=['서울특별시','경기도','대구광역시']
    finaldata_copy.loc[2023:2054].to_csv(f'../data/house_price({size}평).csv')
```
