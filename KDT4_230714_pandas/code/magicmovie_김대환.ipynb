```python
grouped=movieDF.groupby(['배급사'])
gdf=grouped.sum(numeric_only=True)
# gdf
gdf.sort_values('누적매출액',ascending=False)
gdf=movieDF
redf=gdf.reset_index()
# redf
ym=redf['분석월'].unique().tolist()
p5 = movieDF[["배급사","누적매출액"]].groupby("배급사").sum()
p=p5[(p5["누적매출액"] >= p5["누적매출액"].mean())].reset_index()
# fig=plt.figure(figsize=(20,10))
# ax1=fig.add_subplot(1,1,1)
# sns.barplot(x='누적매출액',y='배급사',data=p,ax=ax1)
```

```python
p1 = redf[["배급사","영화명","누적매출액",'분석월','장르']]#.groupby("배급사").sum()
```

```python
# 공동배급사 정리
_x = p1[p1['배급사'].str.contains(",")].copy()
_x

_x["배급사"] = _x["배급사"].apply(lambda x : x.split(","))

spdata=_x.explode("배급사")
good=p1[p1['배급사'].str.contains(',')==0]
```

```python
bestData=pd.concat([good,spdata]).reset_index(drop=True)
```

# 어떤 배급사에서 투자를 받는 것이 좋을까?
---

```python
# 각 배급사가 투자한 영화들의 누적매출액 합계
ranking=bestData.sort_values(axis=0,by='누적매출액',ascending=False).reset_index(drop=True)
ranking['순위']=range(1,len(bestData.values)+1)
ranking['누적매출액(단위:억)']=ranking['누적매출액']/100_000_000
top=ranking[['배급사','영화명','누적매출액']].drop_duplicates(subset=['배급사','영화명'],keep='last').groupby(by='배급사').sum().sort_values(by='누적매출액',ascending=False).reset_index()
top['누적매출액(단위:억)']=top['누적매출액']/10000000
top['순위']=range(1,len(top.values)+1)
top10=top.iloc[0:10].set_index("순위")
fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(1,1,1)
sns.barplot(x='누적매출액(단위:억)',y='배급사',data=top10,ax=ax1,palette="flare")
ax1.set_title('상위 10개 배급사')
#'ch:s=.25,rot=-.25'
```

```python
# 각 배급사가 투자한 영화들의 누적매출액 합계 top 10
top=ranking[['배급사','영화명','누적매출액']].drop_duplicates(subset=['배급사','영화명'],keep='last').groupby(by='배급사').sum().sort_values(by='누적매출액',ascending=False).reset_index()
top['누적매출액(단위:억)']=top['누적매출액']/10000000
top['순위']=range(1,len(top.values)+1)
top.iloc[0:10].set_index("배급사")
```
